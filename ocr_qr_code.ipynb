```python
# create qr code

import qrcode
# example data
data = "qr code for forsorge Patientenverfuegung Hauptblatt 1"
# output file name
filename = "qr_forsorge_hb_1.png"
# generate qr code
img = qrcode.make(data)
# save img to a file
img.save(filename)
```

```python
# read qr code

import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# read the QRCODE image
img = cv2.imread("hauptblatt_1_filled_with_QR_code.JPG")

plt.figure(figsize=(15, 15))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
# initialize the cv2 QRCode detector
detector = cv2.QRCodeDetector()
```

```python
# detect and decode
data, bbox, straight_qrcode = detector.detectAndDecode(img)
```

```python
# if there is a QR code
if bbox is not None:
    print(f"QRCode data:\n{data}")
    # display the image with lines
    # length of bounding box
    n_lines = len(bbox)
    for i in range(n_lines):
        # draw all lines
        point1 = tuple(bbox[i][0])
        point2 = tuple(bbox[(i+1) % n_lines][0])
        cv2.line(img, point1, point2, color=(255, 0, 0), thickness=2)
```

```python
# display the result

# opens image, can only be closed/interrupted on my machine by restarting Kernel
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
